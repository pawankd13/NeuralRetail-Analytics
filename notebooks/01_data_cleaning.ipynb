```python
# dataset info
import pandas as pd

df = pd.read_excel(
    "../data/online_retail_II xlsx kaggle.xlsx"
)

df.head()
```

```python
# missing values
df.isnull().sum()
```

```python
# duplicate record count
df.duplicated().sum()
```

```python
#removing duplicates
df = df.drop_duplicates()
```

### Duplicate Records Handling

The dataset contained 6,865 duplicate records.
These duplicates were removed to ensure accurate sales, customer, and inventory analysis.

```python
df.shape
```

```python
# missing values
df.isnull().sum()
```

```python
# removing missing customer IDs
df = df.dropna(subset=["Customer ID"])
```

```python
df.shape
```

```python
# removing missing descriptions
df = df.dropna(subset=["Description"])
```

```python
df.shape
```

### Handling Missing Values

- Description column contained 2,928 missing values.
- Customer ID column contained 107,833 missing values.

Rows with missing Customer IDs were removed because customer identification is essential for RFM analysis, churn prediction, and customer segmentation.

Rows with missing product descriptions were also removed to maintain data quality.

```python
# negative quantities (returns/cancellations)
(df["Quantity"] < 0).sum()
```

```python
df = df[df["Quantity"] > 0]
```

```python
df.shape
```

```python
# invalid prices
(df["Price"] <= 0).sum()
```

```python
df = df[df["Price"] > 0]
```

```python
df.shape
```

```python
# creating revenue
df["Revenue"] = df["Quantity"] * df["Price"]
```

```python
df.head()
```

```python
# converting invoice date
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

```python
# creating date features
df["Year"] = df["InvoiceDate"].dt.year
```

```python
df["Month"] = df["InvoiceDate"].dt.month
```

```python
df["Day"] = df["InvoiceDate"].dt.day
```

```python
df["Weekday"] = df["InvoiceDate"].dt.day_name()
```

```python
df.shape
```

```python
df.head
```

```python
df.to_csv(
    "../data/cleaned_retail.csv",
    index=False
)
```
